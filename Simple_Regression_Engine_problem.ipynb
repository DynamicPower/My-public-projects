```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2)

# Create a linear regression object
linear_regression_model = linear_model.LinearRegression()

# Load the training data
train_engine_sizes = np.asanyarray(train_data[['ENGINESIZE']])
train_co2_emissions = np.asanyarray(train_data[['CO2EMISSIONS']])

# Train the linear regression model
linear_regression_model.fit(train_engine_sizes, train_co2_emissions)

# Load the testing data
test_engine_sizes = np.asanyarray(test_data[['ENGINESIZE']])
test_co2_emissions = np.asanyarray(test_data[['CO2EMISSIONS']])

# Make predictions on the testing data
predictions = linear_regression_model.predict(test_engine_sizes)

# Print the coefficients and intercept
print('Coefficients:', linear_regression_model.coef_)
print('Intercept:', linear_regression_model.intercept_)

# Evaluate the model
print('Mean squared error: %.2f'
      % mean_squared_error(test_co2_emissions, predictions))
print('Coefficient of determination: %.2f'
      % r2_score(test_co2_emissions, predictions))
```
